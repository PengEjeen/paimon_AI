```python
import json

# JSON 파일에서 데이터 불러오기
with open('./paimonData.json', 'r', encoding='utf-8') as json_file:
    data = json.load(json_file)
print
```

```python
import os
import torch
from datasets import load_dataset
from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    BitsAndBytesConfig,
    TrainingArguments,
    pipeline,
    logging,
)
from peft import LoraConfig
from trl import SFTTrainer
```

```python
#set kakao for train data format
training_data = []

for temp_data in data:
    query = temp_data["instruction"]
    answer = temp_data["output"]

    tmpList = {'text': f"<s>[INST] {query} [/INST] {answer} </s>"}
    training_data.append(tmpList)
```
